```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import os

IMAGES_DIR = os.path.join(os.getcwd(), 'images')
IMG_FILE_PATH = [os.path.join(IMAGES_DIR, file) for file in os.listdir(IMAGES_DIR)]

for i, path in enumerate(IMG_FILE_PATH):
    print(i, path)
```

```python
def showDetail(img):
    print(img.shape, img.dtype, type(img))
```

```python
img1 = plt.imread(IMG_FILE_PATH[1])
showDetail(img1)

plt.imshow(img1)
```

```python
img2 = cv2.imread(IMG_FILE_PATH[3], -1)
showDetail(img2)

cv2.imshow('Image', img2)

cv2.waitKey(0)
cv2.destroyAllWindows()
```
