```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import os
import copy

IMAGES_DIR = os.path.join(os.getcwd(), 'images')
IMG_FILE_PATH = [os.path.join(IMAGES_DIR, file) for file in os.listdir(IMAGES_DIR)]

for i, path in enumerate(IMG_FILE_PATH):
    print(i, path)
```

```python
img = cv2.imread(IMG_FILE_PATH[5], -1)
img1 = img[:, :, : : -1].copy()

img1[:, :, 0] = 255 #R
img[:, :, 0] = 255 #B

plt.imshow(img1)
cv2.imshow('Image', img)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
img = cv2.imread(IMG_FILE_PATH[5], -1)
img = img[:, :, : : -1]

img1 = cv2.cvtColor(img, cv2.COLOR_RGB2BGR)
img2 = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
img3 = cv2.cvtColor(img, cv2.COLOR_RGB2LUV)

plt.figure(figsize = (15, 10))
plt.subplot(221)
plt.imshow(img)
plt.subplot(222)
plt.imshow(img1)
plt.subplot(223)
plt.imshow(img2, 'gray')
plt.subplot(224)
plt.imshow(img3)
```
