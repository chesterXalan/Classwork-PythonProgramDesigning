```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import os
%matplotlib inline

IMAGES_DIR = os.path.join(os.getcwd(), 'images')
IMG_FILE_PATH = [os.path.join(IMAGES_DIR, file) for file in os.listdir(IMAGES_DIR)]

for i, path in enumerate(IMG_FILE_PATH):
    print(i, path)
```

```python
img = cv2.cvtColor(cv2.imread(IMG_FILE_PATH[6], -1), cv2.COLOR_BGR2RGB)
img_gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)

sobel_x = cv2.Sobel(img_gray, -1, 1, 0)
sobel_y = cv2.Sobel(img_gray, -1, 0, 1)
result = cv2.addWeighted(sobel_x, 0.5, sobel_y, 0.5, -1)

plt.figure(figsize = (15, 10))
plt.subplot(221)
plt.imshow(img)
plt.subplot(222)
plt.imshow(sobel_x, 'gray')
plt.subplot(223)
plt.imshow(sobel_y, 'gray')
plt.subplot(224)
plt.imshow(result, 'gray')
```

```python
img = cv2.cvtColor(cv2.imread(IMG_FILE_PATH[7], -1), cv2.COLOR_BGR2RGB)
myFilter = np.ones((7, 7), dtype = 'uint8') / 49
imgFiltered = cv2.filter2D(img, -1, myFilter)

plt.figure(figsize = (15, 10))
plt.subplot(121)
plt.imshow(img)
plt.subplot(122)
plt.imshow(imgFiltered)
```
