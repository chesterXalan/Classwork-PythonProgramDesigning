```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import os
import copy
%matplotlib inline

IMAGES_DIR = os.path.join(os.getcwd(), 'images')
IMG_FILE_PATH = [os.path.join(IMAGES_DIR, file) for file in os.listdir(IMAGES_DIR)]

for i, path in enumerate(IMG_FILE_PATH):
    print(i, path)
```

```python
img = cv2.imread(IMG_FILE_PATH[4], -1)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
imgEq = copy.copy(img)

RGB = ['r', 'g', 'b']
def Histogram(imgInput):
    for i, color in enumerate(RGB):
        imgHist = cv2.calcHist(imgInput, [i], None, [256], [0, 256])        
        plt.plot(imgHist, RGB[i])
        
for color in range(3):
    imgEq[:, :, color] = cv2.equalizeHist(img[:, :, color])        
        
plt.figure(figsize = (15, 10))
plt.subplot(221)
plt.imshow(img)
plt.subplot(222)
Histogram(img)
plt.subplot(223)
plt.imshow(imgEq)
plt.subplot(224)
Histogram(imgEq)
```

```python
img = cv2.imread(IMG_FILE_PATH[5], -1)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
img = (img * 0.5).astype('uint8')

plt.figure(figsize = (15, 5))
plt.subplot(121)
plt.imshow(img)
plt.subplot(122)
Histogram(img)
```
